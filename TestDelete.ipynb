```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define two sets
set1 = {1, 2, 3, 4, 5, 10, 11, 12, 13, 14}
set2 = {3, 4, 5, 6, 7}

# Calculate the intersection and disjoint union
intersection = set1.intersection(set2)
disjoint_union = set1.symmetric_difference(set2)

# Create a Venn diagram
venn2(subsets=(len(set1 - intersection), len(set2 - intersection), len(intersection)),
      set_labels=('Set 1', 'Set 2'))

# Add labels
plt.text(0.2, 0.8, f'Intersection: {len(intersection)}', fontsize=12)
plt.text(0.5, -0.1, f'Disjoint Union: {len(disjoint_union)}', fontsize=12)

# Show the plot
plt.title("Venn Diagram of Sets")
plt.show()
```
